```python
#Retirar acentuação
def corrigir_nomes(nome):
    nome = nome.replace('á', 'a').replace('ç', 'c').replace('ô', 'o').replace('é', 'e').replace('ú', 'u').replace('ó', 'o').replace('ã', 'a').replace('ê', 'e').replace('ú', 'u').replace('í', 'i').replace('r$','')
    return nome

df = df.apply(corrigir_nomes)
df.head()
```

```python
df.columns
```

```python
#Retirar caracteres das colunas numéricas
def corrigir_valores(df,coluna):
    df[coluna] = df[coluna].apply(lambda x: x.replace('r$','').replace(' ','').replace('.','').replace(',','.').replace('-','0'))
    return df[coluna]

corrigir_valores(df, 'vencimentos')
corrigir_valores(df, 'encargos')
corrigir_valores(df, 'beneficios')
corrigir_valores(df, 'outrasremuneracoes')
```

## 1. Definindo a população

- `Vínculo` diferente de desligado;
- `Rendimento total` > 0


A nossa população consiste em todos os colaboradores ativos em 2018. 

### Percepção do erro

- Dado que conhecemos a média da população, o quanto queremos nos distanciar dela ao inferir a média amostral?
- Qual o erro aceitável?

Quanto menor a amostra, maior o erro;

### Cálculo do tamanho da amostra

Precisamos de 3 informações para calcular o tamanho da amostra:

- Valor crítico que depende do grau de confiança:  
        Para esse estudo vamos assumir um grau de confiança de 95% que tem um valor crítico de 1,96;

- O desvio padrão populacional:
        Desvio padrão = 6755,38
        
- Erro máximo de estimativa (e_max):
        Pela nossa percepção o erro máximo são 3 valores a serem testados, 500, 700 e 1000.
        Adotaremos o 700 como erro aceitável. 
        

```python
desvio_padrao = df_ativo['rendimento_total'].std()
desvio_padrao
```

```python
df_ativo.shape
```

```python
#Calculando o tamanho da amostra considerando um erro máximo (e_max) = 500
valor_critico = 1.96
e_max = 500

n = (valor_critico*desvio_padrao/float(e_max))**2
n
```

```python
#Calculando o tamanho da amostra considerando um erro máximo (e_max) = 700
valor_critico = 1.96
e_max = 700

n = (valor_critico*desvio_padrao/float(e_max))**2
n
```

```python
#Calculando o tamanho da amostra considerando um erro máximo (e_max) = 1000
valor_critico = 1.96
e_max = 1000

n = (valor_critico*desvio_padrao/float(e_max))**2
n
```

- Para um erro máximo de 700 precisaremos de uma amostra de 357.78 (360).
